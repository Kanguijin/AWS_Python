```python
csv_data =[]
with open('전국건강증진센터표준데이터_touched.csv') as f:
    for line in f:
        csv_data.append(line.split(','))

csv_data = csv_data[1:]
print(csv_data[:3])
```

```python
cent_count = len(csv_data)
print(cent_count)

doctor_count = 0
doctor_list = []
for i in csv_data:
    if len(i[11]) > 0: 
        doctor_list.append(i[11])

doctor_count = doctor_list.count('0')
print(doctor_count)
     
```

```python
name_list=[]
for e in csv_data[1:]:
    if len(e[2])>0:
        name_list.append(e[2].split()[0])
        #print(e[2].split()[0])
print(name_list)
```

```python
for n in name_list:
    if n[0]=='"':
        print(n[1:])
    else:
        print(n)
```

```python
for i, n in enumerate(name_list):
    name_list[i] = n.lstrip('"')
    #print(n.lstrip('"'))
print(name_list[:10])
```

```python
# lambda x: x[1]
def sort_key(x):
    return x[1]


def unique_count(raw_list, sort = False):
    raw_set = set(raw_list)
    v_list=[]
    for n in raw_set:
        v_list.append(raw_list.count(n)) 

    ret_dict = dict(zip(raw_set, v_list))
    if sort == True:
        ret_dict = dict(sorted(ret_dict.items(), key=lambda x: x[1]))
        #ret_dict = dict(sorted(ret_dict.items(), key=sort_key))
        
    return ret_dict

unique_count(name_list, sort=True)
```
