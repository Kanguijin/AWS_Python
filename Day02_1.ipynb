```python
data_list = []
with open('전국건강증진센터표준데이터.csv') as f:
    for line in f:
        data_list.append(line.split(','))
#print(data_list[1:])
```

```python
si_list = []
for i in data_list[1:]:
    address = i[2]
    address_list = address.split() 
    si_list.append(address_list[1])
print(si_list)

do_list = []
for i in data_list[1:]:
    address = i[2]
    address_list = address.split() 
    do_list.append(address_list[0])
print(do_list)
```

```python
si_set = set(si_list)
d_si_list=[]
d_si_count=[]
for i in si_set:
    d_si_list.append(i)
    d_si_count.append(si_list.count(i))
print(d_si_list)
print(d_si_count)

do_set = set(do_list)
d_do_list=[]
d_do_count=[]
for i in do_set:
    d_do_list.append(i)
    d_do_count.append(do_list.count(i))
print(d_do_list)
print(d_do_count)
```

```python
d1 = dict(zip(d_si_list, d_si_count))
print(d1)
d2 = dict(zip(d_do_list, d_do_count))
print(d2)
```

```python
def fn10(x):
    return x[1]
d1_sorted = dict(sorted(d1.items(), key=fn10))
print(d1_sorted)
```
